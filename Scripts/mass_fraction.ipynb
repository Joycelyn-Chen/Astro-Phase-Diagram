# Energy Map Generation

```python
# Full energy formula
# Energy = (1/2) * m * v_sq + (3/2) k * temp * n

energy_map = yt.YTArray(np.zeros((800, 800, 800))) * yt.units.erg / yt.units.cm**3
cell_volume = 1.0000000000000004*yt.units.pc**3    #2.9379989445851796e+55
k = yt.physical_constants.kb
mu = 1.4
m_H = yt.physical_constants.mass_hydrogen

for i in range(800):
    for j in range(800):
        #m = (obj["flash", "dens"][i,j,0]) * cell_volume
        rho = obj["flash", "dens"][i,j,0]
        v_sq = obj["flash", "velx"][i,j,0]**2 + obj["flash", "vely"][i,j,0]**2 + obj["flash", "velz"][i,j,0]**2
        temp = obj["flash", "temp"][i,j,0]
        n = obj["flash", "dens"][i,j,0] / (mu * m_H)
        energy_map[i][j][0] = ((1/2) * rho * v_sq + (3/2) * k * temp * n).to(energy_map.units)
        
```

```python
plt.imshow(np.log10(energy_map[:,:,0]), cmap = 'RdBu')
plt.colorbar()
plt.savefig("total_energy.png")
```

```python
# Kinetic energy
kinetic_map = yt.YTArray(np.zeros((800, 800, 1))) * yt.units.erg / yt.units.cm**3
# cell_volume = 1.0000000000000004*yt.units.pc**3    #2.9379989445851796e+55
k = yt.physical_constants.kb

for i in range(800):
    for j in range(800):
        rho = obj["flash", "dens"][i,j,0]
        v_sq = obj["flash", "velx"][i,j,0]**2 + obj["flash", "vely"][i,j,0]**2 + obj["flash", "velz"][i,j,0]**2
        kinetic_map[i][j][0] = ((1/2) * rho * v_sq ).to(energy_map.units)
        
        
plt.imshow(np.log10(kinetic_map), cmap = 'RdBu', vmin = -16, vmax = -9)
plt.colorbar()
plt.savefig("kinetic.png")
```

```python
# Thermal energy
thermal_map = yt.YTArray(np.zeros((800, 800, 1))) * yt.units.erg / yt.units.cm**3
k = yt.physical_constants.kb
mu = 1.4
m_H = yt.physical_constants.mass_hydrogen

for i in range(800):
    for j in range(800):
        temp = obj["flash", "temp"][i,j,0]
        n = obj["flash", "dens"][i,j,0] / (mu * m_H)
        thermal_map[i][j][0] = ((3/2) * k * temp * n).to(energy_map.units)
        
        
plt.imshow(np.log10(thermal_map), cmap = 'RdBu', vmin = -16, vmax = -9)
plt.colorbar()
plt.savefig("thermal.png")
```

# Phase Diagram 

```python
# Phase Diagram

# Initialization
datadir = "/Users/joycelynchen/Desktop/UBC/Research/Data/synthetic_data/synthetic_data"
mask_root = "./case_masks_200"
filename_prefix = "sn34_smd132_bx5_pe300_hdf5_plt_cnt_0200"


center =  [0, 0, 0]*yt.units.pc
arb_center = ds.arr(center,'code_length')
left_edge = arb_center - ds.quan(400,'pc')
right_edge = arb_center + ds.quan(400,'pc')

# Loading the data
ds = yt.load(os.path.join(datadir, filename_prefix))
obj = ds.arbitrary_grid(left_edge, right_edge, dims=[800, 800, 800]*yt.units.pc)


single_phase = yt.YTArray(np.zeros((800, 800, 15))) * (yt.units.kelvin) / yt.units.cm**3
dens = yt.YTArray(np.zeros((800, 800, 15))) * (1) / yt.units.cm**3
mu = 1.4
m_H = yt.physical_constants.mass_hydrogen

middle_plane = 300
start = middle_plane - 7
end = middle_plane + 8

for k in range(start, end, 1):
    for i in range(800):
        for j in range(800):
            temp = obj["flash", "temp"][i,j,k]
            n = obj["flash", "dens"][i,j,k] / (mu * m_H)
            dens[i][j][k-start] = (n).to(dens.units)
            single_phase[i][j][k-start] = (n * temp).to(single_phase.units)
    print(f"Whole slice {k} done. [ {round(((k-start)/15)*100, 1)}% ]")
        
single_sum = single_phase.sum(axis=2)
dens_sum = dens.sum(axis=2)

plt.imshow(np.log10(single_sum), cmap = 'RdBu')
plt.colorbar()
plt.savefig("Single_phase.png")
```

```python
# Full version
# Read the mask for each slice and calculate the thermal pressure under the masked area

# Thermal pressure = n * temp
masked_single_phase = yt.YTArray(np.zeros((800, 800, 15))) * (yt.units.kelvin) / yt.units.cm**3
masked_dens = yt.YTArray(np.zeros((800, 800, 15))) * (yt.units.g) / yt.units.cm**3
mu = 1.4
m_H = yt.physical_constants.mass_hydrogen


for z in range(start, end, 1):
    mask_filename = f"{filename_prefix}_z{z}.png"
    mask = cv.imread(os.path.join(mask_root, mask_filename))
    coordinates = np.argwhere(mask == 255)
    for coord in coordinates:
        x, y = coord[0], coord[1]
        
        temp = obj["flash", "temp"][x,y,z]
        n = obj["flash", "dens"][x,y,z] / (mu * m_H)
        masked_dens[i][j][k-start] = (n).to(single_phase.units)
        masked_single_phase[x][y][z-start] = (n * temp).to(masked_single_phase.units)

        
    print(f"Masked region slice {z} done. [ {round(((z-start)/15)*100, 1)}% ]")

masked_single_sum = masked_single_phase.sum(axis=2)
masked_dens_sum = masked_dens.sum(axis=2)

plt.imshow(np.log10(masked_single_sum), cmap = 'RdBu')
plt.colorbar()
plt.savefig("Masked_single_phase.png")
```

```python
# thermal pressure as a function of density in scatter plot
x = np.array(dens_sum.to_ndarray().flatten())
y = np.array(single_sum.to_ndarray().flatten())
x1 = np.array(masked_dens_sum.to_ndarray().flatten())
y1 = np.array(masked_single_sum.to_ndarray().flatten())


# color: gray(#7f7f7f), 
plt.scatter(x, y, c='#1f77b4', marker='o', label='whole slice', s=0.01)
plt.scatter(x, y1, c='r', marker='o', label='masked area', s=0.01)
plt.legend(loc='upper left')

plt.title("Phase Diagram - 1")
plt.xlabel("Density (n)")
plt.ylabel("Thermal pressure (nT)")
plt.grid(True) 
plt.yscale('log')
plt.xscale('log')  
plt.show() 
```
